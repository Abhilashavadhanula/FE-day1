## in two diff data sets we have country_code as common can we combine this two datasets (yes) Merge

```python
final_data=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
ratings=final_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

### observations
1. when rating btwn 4.5 to 4.9 is excellent
2. when ratings are btwn 3.5 to 4.4 is v good
3. if the rating is btw 3.5 to 3.9 is good
4.if the rating is btw 2.5 to 2.9 is avg
5. if the rating is btw 2.0 to 2.4 is poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)


sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings,hue='Rating color',palette=['black','Red','Orange','Yellow','Green','DarkGreen'])
```

## observations

2. Maximum number of rating are between 2.5 to 3.4
